Las clase de Numpy es llamada ndarray y sus atributos son los siguientes:

## **ndarray.ndim**

que nos dice el numero de ejes de la dimensión del arreglo.

Por ejemplo: Hicimos una lista que tiene sólamente 1 eje con 5 elementos, al imprimirlo pues nos dice su dimensión es de 1.



```python
a=[3, 5, 7, 9]
a= np.array(a)
print(a.ndim)
```

##**ndarray.shape**

nos dice las dimensiones del arreglo.

Por ejemplo: Hicimos una lista donde al imprimirla nos dirá el tamaño de la dimensión que es 2.

```python
b=[3, 5]
b=np.array(b)
print(b.shape)
```

##**ndarray.size**

nos dice la cantidad de elementos en la matriz.

Por ejemplo: Hicimos una lista donde tiene dos dimensiones, lo cual su eje vertical tiene 2 y su eje horizontal tiene 2, 2*2=4.

```python
c=([(3, 5),(2,4)])
c=np.array(c)
print(c.size)
```

##**ndarray.dtype**

nos dice el tipo de dato del arreglo.

Ejemplo: La lista tiene enteros y por lo cual nos dirá que es int64.

```python
d=[3, 5]
d=np.array(d)
print(d.dtype)
```

#EJES (AXIS)

Te ayudan a entender dónde están ubicados los datos en el gráfico, es decir, que podemos personalizar esos limites de los ejes bidimensionales.

por ejemplo pondremos una leyenda con el código "plt.legend("Leyenda")":

```python
x=np.linspace(0,2*np.pi,100) #eje x
y=np.sin(x) #eje y

#usar matplotlib

plt.figure(figsize=(8,4)) #dandole la figurita al dibujo #figsize es el tamaño

plt.title("Mi primer gráfico cientifica en PO")

plt.plot(x,y) #plot es graficar el eje x y el eje y en el arreglo

plt.xlabel("x") #nombre del eje x

plt.ylabel("seno de x") #nombre del eje y

plt.grid(True)

plt.legend("Leyenda")
```

##Gráfico de Dispersión

En el siguiente código se realiza un gráfico de dispersión a través de Matplotlib.  

```python
np.random.seed(5)  # Esta línea establece un generador de números aleatorios de numpy.
data = {'a': np.arange(50),
        'c': np.random.randint(0, 15, 50),
        'd': np.random.randn(50)}  #Aquí se crea arreglos
data['b'] = data['a'] + 10 * np.random.randn(50) #operaciones
data['d'] = np.abs(data['d']) * 100 #operaciones

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained') #Se crea la figura y el tamaño de la figura guarados en fig y ax
ax.scatter('a', 'b', c='c', s='d', data=data) #Aqui se anda trazando el gráfico de dispersión, los valores de a es para el eje x, los de b para eje y, los valores de c para el color de los punto, d se utiliza para eltamaño de los puntos
ax.set_xlabel('entry a') #nombre del eje x
ax.set_ylabel('entry b') #nombre del eje y
```

En este tipo de metodo podemos recrear la figura implicitamente con plt.figura que le damos un ancho y altura, ya las siguiente líneas especificamos el nombre de los ejes y creamos una grilla.

La página de guía de matplotlib nos recomienda usar el método OO-style sólo si las figuras son complicadas y para proyectos más grandes.

##my_plotter

Función que ayuda a hacer un gráfico utilizando datos proporcionados.



```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph.
    """
    out = ax.plot(data1, data2, **param_dict)
    return out

data1, data2, data3, data4 = np.random.randn(4, 100)  # make 4 random data sets
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```

##Estilos de línea set_linestyle

Este tipo de grafico sirve para darle estilo a las lineas o funciones, se puede modificar el color, la anchura de línea y estilo de línea.

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
x = np.arange(len(data1))
ax.plot(x, np.cumsum(data1), color='black', linewidth=4, linestyle='--')
l, = ax.plot(x, np.cumsum(data2), color='red', linewidth=2)
l.set_linestyle(':')
```

##Anotaciones annotate()

Se puede usar anotaciones en puntos de un gráfico, conectado por una flecha que apunta con un texto (xytext).

```python
fig, ax = plt.subplots(figsize=(5, 2.7))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = ax.plot(t, s, lw=2)

ax.annotate('aqui hay algo', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='yellow', shrink=0.05))

ax.set_ylim(0, 10)
```

##Escalas! set_yscale o set_xscale

Existen muchos tipos de escalas, se utiliza supuestamente para sólamente la logaritmicas más que todo.



```python
fig, axs = plt.subplots(1, 2, figsize=(5, 2.7), layout='constrained')
xdata = np.arange(len(data1))  # make an ordinal for this
data = 10**data1
axs[0].plot(xdata, data)

axs[1].set_yscale('log')
axs[1].plot(xdata, data)
```

## Graficos utilizando datos generados con Numpy

Se utilizan varios método de trazado como 'pcolormesh()', 'contourf()', 'imshow()' y 'scatter()' para visualizar datos en diferentes formatos.

Es super util para mapas de colores, contornos, investigación de temperaturas del suelos, etc. En resumen, para analizar y visualizar datos.

```python
X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

fig, axs = plt.subplots(2, 2, layout='constrained')
pc = axs[0, 0].pcolormesh(X, Y, Z, vmin=-1, vmax=1, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[0, 0])
axs[0, 0].set_title('pcolormesh()')

co = axs[0, 1].contourf(X, Y, Z, levels=np.linspace(-1.25, 1.25, 11))
fig.colorbar(co, ax=axs[0, 1])
axs[0, 1].set_title('contourf()')

pc = axs[1, 0].imshow(Z**2 * 100, cmap='plasma',
                          norm=mpl.colors.LogNorm(vmin=0.01, vmax=100))
fig.colorbar(pc, ax=axs[1, 0], extend='both')
axs[1, 0].set_title('imshow() with LogNorm()')

pc = axs[1, 1].scatter(data1, data2, c=data3, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[1, 1], extend='both')
axs[1, 1].set_title('scatter()')
```
